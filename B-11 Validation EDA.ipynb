```python
df['Booking_date'] = pd.to_datetime(df['Booking_date'], format='%m/%d/%Y')
df['Expected_checkin'] = pd.to_datetime(df['Expected_checkin'], format='%m/%d/%Y')
df['Expected_checkout'] = pd.to_datetime(df['Expected_checkout'], format='%m/%d/%Y')
```

```python
# Calculate Lead Time
df['Lead_Time'] = (df['Expected_checkin'] - df['Booking_date']).dt.days
```

```python
# calculate Length_of_Stay and convert dates
df['Expected_checkin'] = pd.to_datetime(df['Expected_checkin'])
df['Expected_checkout'] = pd.to_datetime(df['Expected_checkout'])
df['Length_of_Stay'] = (df['Expected_checkout'] - df['Expected_checkin']).dt.days

# revenue and ADR
df['Total_Revenue'] = df['Room_Rate'] * df['Length_of_Stay']
df['ADR'] = df['Total_Revenue'] / df['Length_of_Stay']

# Select columns 
final_table = df[['Expected_checkin',
                  'Expected_checkout',
                  'Length_of_Stay',
                  'Room_Rate',
                  'Total_Revenue',
                  'ADR',
                  'Lead_Time',
                  'Adults',
                  'Children',
                  'Previous_Cancellations']]
final_table.head()
```

```python
# Convert date columns
df["Expected_checkin"] = pd.to_datetime(df["Expected_checkin"])
df["Expected_checkout"] = pd.to_datetime(df["Expected_checkout"])

print(df[["Expected_checkin", "Expected_checkout", "Length_of_Stay"]].head())
```

```python
#calculate booking revenue after applying discount
df["booking_revenue"] = (
    df["Room_Rate"] *
    df["Length_of_Stay"] *
    (1 - df["Discount_Rate"] / 100))

print(df[["Room_Rate", "Discount_Rate", "Length_of_Stay", "booking_revenue"]].head())
```

```python
# Total Potential Revenue
total_revenue = df["booking_revenue"].sum()

print("Total Revenue Potential:", round(total_revenue, 2))


# Filter Cancelled Bookings
cancelled_df = df[df["Reservation_Status"] == 2]

revenue_lost = cancelled_df["booking_revenue"].sum()

print("Cancelled Bookings:", len(cancelled_df))
print("Revenue Lost:", round(revenue_lost, 2))


# Revenue Loss Percentage
if total_revenue != 0:
    revenue_loss_percentage = (revenue_lost / total_revenue) * 100
else:
    revenue_loss_percentage = 0

print("Revenue Loss Percentage:", round(revenue_loss_percentage, 2), "%")


#  Average Revenue per Cancelled Booking
if len(cancelled_df) != 0:
    avg_cancelled_value = cancelled_df["booking_revenue"].mean()
else:
    avg_cancelled_value = 0

print("Average Revenue per Cancelled Booking:", round(avg_cancelled_value, 2))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#cancellation rate by hotel type
TARGET_COL = "Reservation_Status"
HOTEL_COL = "Hotel_Type"  


df["Cancelled"] = (df[TARGET_COL] == 2).astype(int)

# Group by hotel type
cancel_rate = df.groupby(HOTEL_COL)["Cancelled"].mean()

plt.figure(figsize=(6,4))

ax = cancel_rate.plot(kind="bar")

plt.title("Cancellation Rate by Hotel Type")
plt.ylabel("Cancellation Rate")
plt.xticks(rotation=45)


for i in range(len(cancel_rate)):
    value = cancel_rate.iloc[i]
    ax.text(
        i,                      
        value,                  
        f"{value:.2%}",         
        ha='center',
        va='bottom'
    )

plt.tight_layout()
plt.show()
```

Bivariate analysis

```python
#Cancellation rate by deposit type
deposit_cancel = (
    df.groupby("Deposit_type")["Cancelled"]
    .agg(["mean", "count"])
    .sort_values("mean", ascending=False)
)

plt.figure(figsize=(7,4))
ax = sns.barplot(x=deposit_cancel.index, y=deposit_cancel["mean"],palette="magma")

plt.title("Cancellation Rate by Deposit Type")
plt.xlabel("Deposit Type")
plt.ylabel("Cancellation Rate")

for i, v in enumerate(deposit_cancel["mean"]):
    ax.text(i, v, f"{v:.1%}", ha="center", va="bottom")

plt.tight_layout()
plt.show()

display(deposit_cancel)
```

```python
#Cancellation rate by hotel type
hotel_cancel = (
    df.groupby("Hotel_Type")["Cancelled"]
    .agg(["mean", "count"])
    .sort_values("mean", ascending=False)
)

plt.figure(figsize=(7,4))
ax = sns.barplot(x=hotel_cancel.index, y=hotel_cancel["mean"],palette="Set2")
plt.title("Cancellation Rate by Hotel Type")
plt.xlabel("Hotel Type")
plt.ylabel("Cancellation Rate")

for i, v in enumerate(hotel_cancel["mean"]):
    ax.text(i, v, f"{v:.1%}", ha="center", va="bottom")

plt.tight_layout()
plt.show()

display(hotel_cancel)
```

Multivariate Analysis 

```python
#Revenue loss by hotel type
cancelled_df = df[df["Cancelled"] == 1].copy()

loss_by_hotel = (
    cancelled_df.groupby("Hotel_Type")["booking_revenue"]
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(7,4))
ax = sns.barplot(x=loss_by_hotel.index, y=loss_by_hotel.values,palette="viridis")
plt.title("Revenue Loss by Hotel Type")
plt.xlabel("Hotel Type")
plt.ylabel("Revenue Loss")

for i, v in enumerate(loss_by_hotel.values):
    ax.text(i, v, f"{v:,.0f}", ha="center", va="bottom")

plt.tight_layout()
plt.show()

print(loss_by_hotel)
```

```python
#Monthly cancellation rate

df["Checkin_Month"] = df["Expected_checkin"].dt.month_name()

month_order = ["January","February","March","April","May","June",
               "July","August","September","October","November","December"]

monthly_cancel = (
    df.groupby("Checkin_Month")["Cancelled"]
    .mean()
    .reindex(month_order)
)

plt.figure(figsize=(10,4))
sns.lineplot(x=monthly_cancel.index, y=monthly_cancel.values, marker="o")
plt.title("Monthly Cancellation Rate")
plt.xlabel("Month")
plt.ylabel("Cancellation Rate")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print(monthly_cancel)
```

```python
#Correlation matrix of key features
corr_cols = ["Cancelled", "Lead_Time", "Length_of_Stay", "Room_Rate",
             "Discount_Rate", "Adults", "Children", "booking_revenue"]

corr = df[corr_cols].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", center=0, linewidths=0.5)
plt.title("Correlation Matrix of Key Features")
plt.tight_layout()
plt.show()
```

```python
status_labels = {
    1: "Check-In",
    2: "Canceled",
    3: "No-Show"
}

df['Reservation_Label'] = df['Reservation_Status'].map(status_labels)

plt.figure(figsize=(6,4))

ax = sns.countplot(
    x="Reservation_Label",
    data=df,
    order=["Check-In","Canceled","No-Show"],
    palette={"Check-In":"green","Canceled":"red","No-Show":"yellow"}
)

total = len(df)

for p in ax.patches:
    percentage = 100 * p.get_height() / total
    ax.text(
        p.get_x() + p.get_width()/2,
        p.get_height(),
        f'{percentage:.1f}%',
        ha='center',
        va='bottom'
    )

plt.title("Reservation Status Distribution")
plt.xlabel("Reservation Status")
plt.ylabel("Number of Bookings")
plt.tight_layout()
plt.show()
```

```python
deposit_cancel = (
    df.groupby("Deposit_type")["Cancelled"]
    .agg(["mean", "count"])
    .sort_values("mean", ascending=False)
)

lowest_rate = deposit_cancel["mean"].min()

plt.figure(figsize=(7,4))
ax = sns.barplot(x=deposit_cancel.index, y=deposit_cancel["mean"],palette="magma")

plt.title("Cancellation Rate by Deposit Type")
plt.xlabel("Deposit Type")
plt.ylabel("Cancellation Rate")

for i, v in enumerate(deposit_cancel["mean"]):
    diff = v - lowest_rate
    ax.text(
        i,
        v,
        f"{v:.2%}\n(+{diff:.2%})" if diff > 0 else f"{v:.2%}\n(lowest)",
        ha="center")
```
